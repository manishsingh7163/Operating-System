```python
def FIFO():
    capacity=int(input("Enter the number of frames: "))
    l,fault,top,pf=[],0,0,'N'
    rs=list(map(int,input("Enter the reference string: ").strip().split()))
    print("\nString ",end='')
    for i in range(capacity):
        print("f%d"%i,end=' ')
    print("Fault\n")
    for i in rs:
        if i not in l:
            if len(l)<capacity:
                l.append(i)
            else:
                l[top]=i
                top=(top+1)%capacity
            fault=fault+1
            pf='Y'
        else:
            pf='N'
        print("  %d\t"%i,end='')
        for x in l:
            print(x,"",end=' ')
        for x in range(capacity-len(l)):
            print('  ',end=' ')
        print(" %s"%pf)
    print("\nTotal no of requests: %d\nTotal Page Faults: %d\nFault Rate: %0.2f%%"%(len(rs),fault,(fault/len(rs))*100))

def LRU():
    capacity = int(input("Enter the number of frames: "))
    l,st,fault,pf=[],[],0,'N'
    rs = list(map(int,input("Enter the reference string: ").strip().split()))
    print("\nString ",end='')
    for i in range(capacity):
        print("f%d"%i,end=' ')
    print("Fault\n")
    for i in rs:
        if i not in l:
            if len(l)<capacity:
                l.append(i)
                st.append(len(l)-1)
            else:
                ind = st.pop(0)
                l[ind] = i
                st.append(ind)
            pf='Y'
            fault=fault+1
        else:
            st.append(st.pop(st.index(l.index(i))))
            pf='N'
        print("  %d\t"%i,end='')
        for x in l:
            print(x,"",end=' ')
        for x in range(capacity-len(l)):
            print('  ',end=' ')
        print(" %s"%pf)
    print("\nTotal Requests: %d\nTotal Page Faults: %d\nFault Rate: %0.2f%%"%(len(rs),fault,(fault/len(rs))*100))



def LFU():
    capacity=int(input("Enter the number of frames: "))
    l,m,fault,pf=[],[],0,'N'
    rs=list(map(int,input("Enter the reference string: ").strip().split()))
    cl={}
    for i in rs:
        cl[i]=0
    print(cl)
    print("\nString ",end='')
    for i in range(capacity):
        print("f%d"%i,end=' ')
    print("Fault\n")
    for i in rs:
        if i not in l:
            if len(l)<capacity:
                l.append(i)
                cl[i]=cl[i]+1
            else:
                for j in l:
                    m.append(cl[j])
                ind=m.index(min(m))
                cl[l[ind]]=0
                l[ind]=i
                cl[i]=cl[i]+1
            pf='Y'
            fault=fault+1
        else:
            cl[i]=cl[i]+1
            pf='N'
        print("  %d\t"%i,end='')
        for x in l:
            print(x,"",end=' ')
        for x in range(capacity-len(l)):
            print('  ',end=' ')
        print(" %s"%pf)
    print("\nTotal Requests: %d\nTotal Page Faults: %d\nFault Rate: %0.2f%%"%(len(rs),fault,(fault/len(rs))*100))

print("FIFO Page Replacement:\n")
FIFO()
print("\nLRU Page Replacement:\n")
LRU()
print("\nLFU Page Replacement:\n")
LFU()
```
