```python
at=[]
bt=[];pq=[];pq1=[];pq2=[];p=[];time=[]
n=int(input("enter no of processes: "))
print("Enter Arrival Time: ")
for i in range (0,n):
    ele=int(input())
    at.append(ele)
print("Enter Burst Time: ")
for i in range (0,n): 
    ele1=int(input())
    bt.append(ele1)
print("Enter Priority Queue values: ")
for i in range (0,n):
    ele=int(input())
    pq.append(ele)
ct=[0,0,0,0,0]
tat=[];wt=[];c=at[0];k=0
for i in range (0,n):
    for j in range (k,n):
        if at[j]<=c:
            if pq[j]==1:
                pq1.append(j)
            elif pq[j]==2:
                pq2.append(j)
            if j==n-1:
                k=n
        else:
            k=j
            break
    if pq1!=[]:
        c=c+bt[pq1[0]]
        ct[pq1[0]]=c
        p.append(pq1[0])
        time.append(c)
        pq1.remove(pq1[0])
    else:
        c=c+bt[pq2[0]]
        ct[pq2[0]]=c
        p.append(pq2[0])
        time.append(c)
        pq2.remove(pq2[0])
        
for i in range (0,n):
    tat.append(ct[i]-at[i])
    wt.append(tat[i]-bt[i])
print("Process id \t Priority \t Arrival time \t Burst time \t CT \t TAT \t WT")
w=0
t=0
for i in range (0,n):
    print(i,"\t\t",pq[i],"\t\t",at[i],"\t\t",bt[i],"\t\t",ct[i],"\t",tat[i],"\t",wt[i])

for i in range (0,len(p)):
    print("p",p[i],"\t",end=" ")
print()
for i in range (0,len(p)):
    print(time[i],"\t",end=" ")    
avg_tat=sum(tat)/5
avg_wt=sum(wt)/5
print("\naverage waiting time: ",avg_wt)
print("average turn around time: ",avg_tat)

```
