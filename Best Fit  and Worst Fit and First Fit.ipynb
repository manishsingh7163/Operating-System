```python
def first_fit(process,memory):
        print("             FIRST FIT           ")
        print("process SIZES","              ","BLOCK-NO")
        m=len(memory)
        n=len(process)
        flag_m=[0]* m           
        flag_f=[0]* n                 
        for i in range(n):
                for j in range(m):
                        if flag_m[j]==0 and memory[j]>=process[i] and process[i]>0:
                                flag_m[j]=1
                                flag_f[i]=1
                                print(process[i],"                              ",j+1)
                                process[i]=0

        for i in range(n):
                if flag_f[i]==0:
                        print("file",i+1,"could'nt be allocated to any block ","\n")
                               
                                        
def best_fit(process,memory):
        print("         BEST FIT          ")
        print("process SIZES","              ","BLOCK-NO")
        m=len(memory)
        n=len(process)
        flag_m=[0]* m                
        flag_f=[0]* n                 
        for i in range(n):
                best=[ ]
                for j in range(m):
                        if flag_m[j]==0 and memory[j]>=process[i] and process[i]>0:
                                best.append(memory[j])
                        elif  memory[j]<process[i] or flag_m[j]==1:
                                best.append(10**4)
                                
                index=best.index(min(best))
                
                flag_m[index]=1
                flag_f[i]=1
                print(process[i],"                              ",index+1)
                process[i]=0


        for i in range(n):
                if flag_f[i]==0:
                        print("file",i+1,"could'nt be allocated to any block ","\n")
        
def worst_fit(process,memory):
        print("            WORST FIT              ")
        print("process SIZES","              ","BLOCK-NO","              ")
        m=len(memory)
        n=len(process)
        flag_m=[0]* m                
        flag_f=[0]* n       
        for i in range(n):
                worst=[ ]
                for j in range(m):
                        if flag_m[j]==0 and memory[j]>=process[i] and process[i]>0:
                                worst.append(memory[j])
                        elif  memory[j]<process[i] or flag_m[j]==1:
                                worst.append(10**(-4))
                                
                if sum(worst)!=(10**(-4))*m:
                        index=worst.index(max(worst))
                        flag_m[index]=1
                        flag_f[i]=1
                        print(process[i],"                              ",index+1)
                        process[i]=0

        for i in range(n):
                if flag_f[i]==0:
                        print("file",i+1,"could'nt be allocated to any block \n")

                        
memory= [100, 500, 200, 300, 600]                   
process=[212, 417, 112, 426]           
x=tuple(process)
process2=list(x)
process3=list(x)

first_fit(process,memory)
best_fit(process2,memory)
worst_fit(process3,memory)
```
