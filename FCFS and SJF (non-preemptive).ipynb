```python
def FCFS(n, br_time):
    wt = 0
    Gchart = []
    for i in range(0, n):
        for j in range(0, br_time[i][1]):
            Gchart.append(br_time[i][0])
        wt += br_time[i][1]*(n-i-1)
    wt /= n
    return Gchart, wt

def SJFNonPreemptive(n, br_time):
    swapList = []
    for i in range(0, n):
        c = i
        for j in range(i+1, n):
            if br_time[j][1]<br_time[c][1]:
                c = j
        if c != i:
            swapList         = br_time[c]
            br_time[c] = br_time[i]
            br_time[i] = swapList
    return FCFS(n, br_time)





def EnterBurstTime(br_time, n):
    for i in range(0, n):
        x = int(input("Enter the burst time for process "+str(i+1)+" : "))
        br_time[i].append(x)

def EnterArrivalTime(ar_time, n):
    for i in range(0, n):
        x = int(input("Enter the arrival time for process "+str(i+1)+" : "))
        ar_time[i].append(x)

def EnterPriority(pr_time, n):
    for i in range(0, n):
        x = int(input("Enter the priority for process "+str(i+1)+" : "))
        pr_time[i].append(x)

def GanttOutput(Gchart):
    firstLine = "|"
    aboveLine = "_"
    underLine = "‾"
    secondLine = "0"
    for i in range(0, len(Gchart)):
        firstLine = firstLine + "P" + str(Gchart[i]) + "|"
        if i<10:
            secondLine = secondLine + "  " + str((i+1))
        else:
            secondLine = secondLine + " " + str((i+1))
    for i in range(1, len(firstLine)):
        underLine += "‾"
        aboveLine += "_"
    return aboveLine + "\n" + firstLine + "\n" + underLine + "\n" + secondLine



while(1):
    print ("\n\nChoose the type of the scheduler:")
    print ("F  for FCFS")
    print ("S  for SJF non preemptive")
    print ("Q  for Quit")
    scheduler = input()
    if scheduler == "Q" or scheduler == 'q':
        break
    n = int(input("Enter the number of processes: "))
    br_time = []
    for i in range(0, n):
        newList = []
        newList.append((i+1))
        br_time.append(newList)
    wt = 0
    Gchart = []
    if scheduler == 'F' or scheduler == 'f':
            EnterBurstTime(br_time, n)
            Gchart, wt = FCFS(n, br_time)
    elif scheduler == 'S' or scheduler == 's':
            EnterBurstTime(br_time, n)
            Gchart, wt = SJFNonPreemptive(n, br_time)
   
    print("Gchart :")
    print(GanttOutput(Gchart))
    print("waiting time = ", wt)
    print("Average time = ", (wt/n))
```
